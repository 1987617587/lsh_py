```python
import numpy as np
import datetime
```

## 1.加载数据

```python
filename = 'data/presidential_polls.csv'

data_arr=np.loadtxt(filename,
               dtype=str,
               delimiter=',',
               skiprows=1,
                usecols=(7,17,18)
                   )
print(data_arr)
```

## 2.数据处理

###  日期数据处理

```python
enddate_ls = data_arr[:,0]
print(enddate_ls)
```

### 时间格式转换

```python
print(datetime.datetime.strptime('9/22/2016','%m/%d/%Y'))
date_ls = [datetime.datetime.strptime(enddate,'%m/%d/%Y') for enddate in enddate_ls]
print(date_ls)
```

```python
month_list = ['%d-%02d'%(date_obj.year,date_obj.month) for date_obj in date_ls]
print(month_list)
```

### list ==>ndarray

```python
month_array = np.array(month_list)
print(month_array)
```

### 对日期数据去重排序

```python
months = np.unique(month_list)
print(months)
```

## 希拉里民调数据处理

### 获取希拉里民调数据

```python
adjpoll_clienton_data = data_arr[:,1]
print(adjpoll_clienton_data)
```

### 数据清洗，处理无效或缺失的数据

```python
adjpoll_clienton_data = np.where(adjpoll_clienton_data == "","0",adjpoll_clienton_data)
print(adjpoll_clienton_data)
```

### 数据格式转换 str==>float64

```python
adjpoll_clienton_data = adjpoll_clienton_data.astype(np.float64)
print(adjpoll_clienton_data)
print(adjpoll_clienton_data.dtype)
```

## 川普民调数据处理

```python
adjpoll_trump_data = data_arr[:,2]
print(adjpoll_trump_data)
```

### 数据清洗，处理无效或缺失的数据

```python
adjpoll_trump_data = np.where(adjpoll_trump_data == "","0",adjpoll_trump_data)
print(adjpoll_trump_data)
```

### 数据格式转换 str==>float64

```python
adjpoll_trump_data = adjpoll_trump_data.astype(np.float64)
print(adjpoll_trump_data)
print(adjpoll_trump_data.dtype)
```

# 3.数据分析

```python
result = []
for month in months:
    codition = month_array == month
    # 当前月份希拉里民调数据的和
    adjpoll_clienton_month_data = adjpoll_clienton_data[codition]
    # 统计求和
    adjpoll_clienton_month_sum = np.sum(adjpoll_clienton_month_data)
    # 当前月份川普民调数据的和
    adjpoll_trump_month_data = adjpoll_trump_data[codition]
    # 统计求和
    adjpoll_trump_month_sum = np.sum(adjpoll_trump_month_data)
    result.append([month,adjpoll_clienton_month_sum,adjpoll_trump_month_sum])
    
```

### 查看两位竞选人每月的统计和

```python
print(result)
```
