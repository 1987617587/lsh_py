# 柱形图

```python
x = np.arange(5)
y1,y2 = np.random.randint(1,25,size=(2,5))
width = 0.33 # 柱形宽度
ax = plt.subplot(1,1,1)
ax.bar(x,y1,width,color='r')
ax.bar(x+width,y2,width,color='#0000ff')
plt.show()
```

# 水平柱形图

```python
mpl.rcParams['font.sans-serif'] = ['SimHei']  # 避免中文乱码的情况
# 构件数
x_data = ['2011','2012','2013','2014','2015','2016','2017','2018']
y_data1 = [58000,60200,63000,71000,84000,86000,91000,96000]
y_data2 = [52000,54000,61000,63000,64300,67000,69000,70700]
bar_height =0.3
y = np.arange(len(x_data))
plt.barh(y=y,width=y_data1,height=bar_height,label='Java语言课程',color='steelblue',alpha=0.8)
plt.barh(y=y+bar_height,width=y_data2,height=bar_height,label='c语言课程',color='indianred',alpha=0.8)
# 设置图形标题
plt.title("Java与C对比")
# 设置坐标轴名称
plt.xlabel('销量')
plt.ylabel("年份")
# 显示图例
plt.legend()
# 在柱形图上显示具体数据
# plt.text():显示文本
# 参数1，参数2：设置文本显示的位置
# 参数3：设置显示的内容
# 参数ha：控制水平对齐方式
# 参数va：控制垂直对齐方式
for y,x in enumerate(y_data1):
    plt.text(x+5000,y-bar_height/2,'%s'%x,ha='center',va='bottom')
for y,x in enumerate(y_data2):
    plt.text(x+5000,y+bar_height/2,'%s'%x,ha='center',va='bottom')
plt.show()
```

# 饼形图

```python
sizes = [60,90,220,180]
labels = ['第一季度','第二季度','第三季度','第四季度']
colors =  ['r','g','b','y']
explode =(0.15,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,labeldistance=1.1,autopct='%3.1f%%',shadow=True,startangle=90,pctdistance=0.6)
plt.show()
```

# 颜色标记线型

```python
fig,axes = plt.subplots(2)
axes[0].plot(np.random.randint(0,100,50),color='r',linestyle='dashed',marker='o') #,alpha=0.5
axes[0].plot(np.random.randint(0,100,50),color='#33aa88',linestyle='dotted',marker='.')
axes[0].plot(np.random.randint(0,100,50),color=(0.5,0.6,0.3),linestyle='solid',marker='>')

axes[1].plot(np.random.randint(0,100,50),'ro--')
```

# 刻度、标签、图例

```python
fig,ax = plt.subplots(1)
ax.plot(np.random.randn(1000).cumsum(),label='line0')
ax.plot(np.random.randn(1000).cumsum(),label='line1')
ax.plot(np.random.randn(1000).cumsum(),label='line2')
# 设置标题
ax.set_title('演示案例')
# 设置坐标轴标签
ax.set_xlabel('Number')
ax.set_ylabel('Month')
# 设置刻度范围
ax.set_xlim([0,800])
#ax.set_ylim([0,30])
# 设置刻度位置
ax.set_xticks(range(0,800,100))
ax.set_yticks(range(0,45,15))
# 设置刻度标签
ax.set_yticklabels(['Jan','Feb','Mar'])
ax.legend(loc='best')
plt.show()
```
