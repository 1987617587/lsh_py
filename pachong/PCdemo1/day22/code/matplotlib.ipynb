```python
x = [1,2,3,4]
y = [2,4,5,9]
plt.scatter(x,y)
plt.show()
```

### 柱形图bar()

```python
x = np.arange(5)
y1,y2 = np.random.randint(1,25,size=(2,5))
width = 0.33 # 柱形宽度
ax = plt.subplot(1,1,1)
ax.bar(x,y1,width,color='r')
ax.bar(x+width,y2,width,color='#0000ff')

plt.show()
```

## 水平柱形图

```python
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 避免中文乱码的情况
# 构建数
x_data = ['2011','2012','2013','2014','2015','2016','2017','2018',]
y_data1 = [58000,60200,62312,70212,81321,86000,91000,98825]
y_data2 = [54000,58000,62000,73202,85321,88000,96000,96825]
bar_height = 0.3
y = np.arange(len(x_data))
plt.barh(y=y,width=y_data1,height=bar_height,label='Java语言课程',color='#0000ff',alpha=0.8)
plt.barh(y=y+bar_height,width=y_data2,height=bar_height,label='C语言课程',color='indianred',alpha=0.8)
plt.title('Java与C对比')
# 设置坐标轴的名称
plt.xlabel('销量')
plt.ylabel('年份')
# 显示图例
plt.legend()
# 在柱形图上显示具体数据

for y,x in enumerate(y_data1):
    # plt.text()：显示文本
    # 参数1，参数2 设置文本的显示位置
    # 参数3 设置显示的内容
    # 参数ha:控制水平的对齐方式
    # 参数va:控制垂直的对齐方式
    plt.text(x+5000,y-bar_height/2,'%s'%x,ha='center',va='bottom')
# 显示y_data2
for y,x in enumerate(y_data2):
    plt.text(x+5000,y+bar_height/2,'%s'%x,ha='center',va='bottom')
plt.show()

```

## 直方图

## 饼形图

```python
sizes = [60,90,220,180]
labels = ['第一季度','第二季度','第三季度','第四季度']
colors = ['r','g','b','y']
explode = (0.15,0,0,0)
patches,l_text,p_text = plt.pie(sizes,explode=explode,labels=labels,colors=colors,
                                labeldistance = 1.1,autopct = '%3.1f%%',
                                shadow = False,startangle = 90,pctdistance = 0.6)

plt.show()
```

## 颜色、标记、线型

```python
fig,axes = plt.subplots(2)  # alpha透明度
axes[0].plot(np.random.randint(0,100,50),color='r',linestyle='dashed',marker='o',alpha=0.8)
axes[0].plot(np.random.randint(0,100,50),color='#33aa88',linestyle='dotted',marker='.')
axes[0].plot(np.random.randint(0,100,50),color=(0.5,0.6,0.6),linestyle='dotted',marker='>')
# 简写方式
axes[1].plot(np.random.randint(0,100,50),'ro-')
```

## 刻度、标签、图例

```python
fig, ax = plt.subplots(1) 
ax.plot(np.random.randn(1000).cumsum(), label='line0')
ax.plot(np.random.randn(1000).cumsum(), label='line1') 
ax.plot(np.random.randn(1000).cumsum(), label='line2')
# 设置标题
ax.set_title("演示案例")
# 设置坐标轴标签
ax.set_xlabel("Number")
ax.set_ylabel("Month")
# 设置刻度范围
ax.set_xlim([0,800])
ax.set_ylim([0,30])
# 设置刻度显示位置
ax.set_xticks(range(0,500,100))
ax.set_yticks(range(-30,30,10))

# 设置刻度标签
ax.set_yticklabels(['Jan','Feb','Mar'])


ax.legend(loc='best') 
plt.show()
```
