```python
import numpy as np
filename = "data/douban.csv"

arr=np.loadtxt(open(filename, encoding='utf8'),
               dtype=str,
               delimiter=',',
               skiprows=1)
print(arr)
```

```python
score_arr = arr[:,1]
print(score_arr)
print(type(score_arr))
print(score_arr.dtype)
```

```python
score_arr = score_arr.astype(np.float64)
print(score_arr)
print(type(score_arr))
print(score_arr.dtype)
```

## 计算平均分

```python
avg_score = np.mean(score_arr)
print(avg_score)
```

## 计算评分最高的电影

```python
max_scroe = np.max(score_arr)
print(max_scroe)
```

```python
## 使用条件索引
```

```python
bool_arr = score_arr==max_scroe
print(bool_arr)
```

```python
max_scroe_movies = arr[bool_arr]
print('评分最高的电影：')
print(max_scroe_movies)
```

```python
## 或者
print(arr[score_arr==max_scroe])
```

## 评分在9分以上的电影


```python
print(arr[score_arr>=9])
```

```python
bool_arr = score_arr>=9
print(bool_arr)
movies = arr[bool_arr]
print('评分在9分以上的电影：')
print(movies)
```
