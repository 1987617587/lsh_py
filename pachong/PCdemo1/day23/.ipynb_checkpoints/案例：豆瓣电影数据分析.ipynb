```python
'''
案例：
    a、用numpy加载csv数据
    b、把评分列转换成float64类型
    *c、计算电影的平均评分
    *d、求评分最高的电影
    *e、求评分在9分以上的电影

'''
```

```python
import numpy as np
```

# a、用numpy加载csv数据

```python
csv_filename = './data/douban.csv'
data_arr = np.loadtxt(csv_filename,delimiter=',',dtype=str,skiprows=1,encoding='utf-8')
print(data_arr)
```

# b、把评分列转换成float64类型

```python
score_arr = data_arr[:,1]
print(score_arr)
print(type(score_arr))
print(score_arr.ndim)
print(score_arr.shape)
print(score_arr.dtype)
```

```python
score_arr = score_arr.astype(np.float64)
print(score_arr)
print(score_arr.dtype)
```

# c、计算电影的平均评分

```python
avg_score = np.mean(score_arr)
print(avg_score)
```

# d、求评分最高的电影

```python
max_score = np.max(score_arr)
print(max_score)
```

```python
bool_arr  = score_arr == max_score
print(bool_arr)
```

```python
max_score_movies = data_arr[bool_arr]
print('评分最高的电影：')
print(max_score_movies)
```

```python
print(data_arr[score_arr == max_score])
```

# e、求评分在9分以上的电影

```python
bool_arr = score_arr >= 9
print(bool_arr)
```

```python
print('评分在9分以上的电影：')
print(data_arr[bool_arr])
```
